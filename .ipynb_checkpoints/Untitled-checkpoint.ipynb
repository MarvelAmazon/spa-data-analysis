```python
def_temp = spa_df_final[spa_df_final[yes_no_col[0]] ==1.0][[yes_no_col[0],'DEPARTN','FACIL']]
def_temp[yes_no_col[0]]= yes_no_col[0]
def_temp.rename(columns = {yes_no_col[0]: 'column_name'},inplace =True)
def_temp.groupby(['column_name','DEPARTN']).size().to_frame().reset_index().rename(columns={0:'count'})
```

```python
map_dict = dict(column_name =[],departn = [],values =[])
for col in yes_no_col:
    val = spa_df_final[spa_df_final[col] ==1.0][[col,'DEPARTN','FACIL']]
    val[col]= col
    val.rename(columns = {col: 'column_name'},inplace =True)
    val=val.groupby(['column_name','DEPARTN']).size().to_frame().reset_index().rename(columns={0:'count'})
    map_dict['column_name'] += list(val['column_name'])
    map_dict['departn'] += list(val['DEPARTN'])
    map_dict['values'] += list(val['count'])
```

```python
for col in col_num:
    val = spa_df_final[[col,'DEPARTN']]
    val['column_name']= col
    val.rename(columns = {col: 'num'},inplace =True)
    val=val.groupby(['column_name','DEPARTN'])['num'].sum().to_frame().reset_index().rename(columns={0:'count'})
    map_dict['column_name'] += list(val['column_name'])
    map_dict['departn'] += list(val['DEPARTN'])
    map_dict['values'] += list(val['num'])
```

```python
for col in yes_no_col:
    val = spa_df_final[spa_df_final[col].isin([1.0,2.0,3.0])][[col,'DEPARTN','FACIL']]
    val[col]= col
    val.rename(columns = {col: 'column_name'},inplace =True)
    val=val.groupby(['column_name','DEPARTN']).size().to_frame().reset_index().rename(columns={0:'count'})
    map_dict['column_name'] += list(val['column_name'])
    map_dict['departn'] += list(val['DEPARTN'])
    map_dict['values'] += list(val['count'])
```

```python
df =pd.DataFrame(map_dict)
```
